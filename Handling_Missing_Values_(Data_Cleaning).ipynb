```python
import pandas as pd
import numpy as np
```

```python
#dictionary of list

dict = {'Marks':[100,90,np.nan,95],
        'Rank':[1,2,4,np.nan],
        'Avg Marks':[np.nan,40,80,98]}
dict
```

```python
df = pd.DataFrame(dict)
df
```

```python
# Checking Null Values
df.isnull() # For Null Values result is true
```

```python
# Fill null values with 0
df.fillna(0)
```

```python
# Fill null values with next value
df.fillna(method='bfill')
```

```python
# Fill null values with prev value
df.fillna(method='pad')
```

```python
# Fill null value with Mean of prev and next value
df
```

```python
# As we can see in Marks column 1 and 3 values are 90 and 95 and mean of this values are 92.5
df.interpolate() # In avg marks there is no prev no value present hence it remains NaN only
```

```python
# Drop null values ( It will delete entire row)

df.dropna()
```
